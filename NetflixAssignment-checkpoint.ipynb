Task 2: Univariate Analysis

```python
# Plot the distribution of: age, watch_hours, monthly_fee, hurned (0 = No churn, 1 = Churn)
numerical_cols = ['age', 'watch_hours', 'monthly_fee', 'churned']
for col in numerical_cols:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Count plots for categorical variables: subscription_type, gender, region, device, payment_method, favorite_genre
for col in categorical_cols:
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.show()
```

Task 3: Bivariate Analysis

```python
# Compare average watch_hours and monthly_fee across by subscription_type, region, device
for feature in ['subscription_type', 'region', 'device']:
    sns.barplot(x=feature, y='watch_hours', data=df, errorbar=None)
    plt.title(f"Average Watch Hours by {feature}")
    plt.show()

    sns.barplot(x=feature, y='monthly_fee', data=df, errorbar=None)
    plt.title(f"Average Monthly Fee by {feature}")
    plt.show()
```

```python
# Plot average avg_watch_time_per_day by favorite_genre.
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df, errorbar=None)
plt.title("Avg Watch Time per Day by Favorite Genre")
plt.show()
```

```python
# Analyze churn rate based on: gender, region, subscription_type, payment_method
for feature in ['gender', 'region', 'subscription_type', 'payment_method']:
    churn_rate = df.groupby(feature)['churned'].mean().reset_index()
    sns.barplot(x=feature, y='churned', data=churn_rate)
    plt.title(f"Churn Rate by {feature}")
    plt.ylabel("Churn Rate")
    plt.show()
```
